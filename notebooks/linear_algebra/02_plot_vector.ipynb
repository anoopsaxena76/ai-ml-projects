```python
import numpy as np
import matplotlib.pyplot as plt

# Define vectors
u = np.array([2, 1])
v = np.array([1, 3])
uv = u + v

# Plot vectors
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='r', label='u = [2,1]')
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='g', label='v = [1,3]')
plt.quiver(0, 0, uv[0], uv[1], angles='xy', scale_units='xy', scale=1, color='b', label='u+v = [3,4]')

# Axes setup
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.gca().set_aspect('equal')

# Title and legend
plt.title("Vector Addition: u + v")
plt.legend()
plt.show()
```
